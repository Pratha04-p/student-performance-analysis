```python
import pandas as pd

df = pd.read_csv("Student_Marks.csv")

df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["time_study"], df["Marks"])
plt.xlabel("Study Time")
plt.ylabel("Marks")
plt.title("Study time vs Marks")
plt.show()
```
